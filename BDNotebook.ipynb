# Limpieza de datos

## Eliminacion de columnas

```python
csvfile.drop(['seguridad.grupo_actor','seguridad.estado_civil','seguridad.actividad_delictiva','seguridad.parentesco','seguridad.ocupacion','seguridad.discapacidad','seguridad.grupo_especial','seguridad.nivel_academico','seguridad.testigo','seguridad.conducta','seguridad.caracterizacion','seguridad.articulo_penal','seguridad.categoria_penal','seguridad.modelo','seguridad.color','seguridad.permiso','seguridad.unidad_medida'], inplace=True, axis=1)
```

# Extraccion de datos de la fecha.

```python
csvfile['seguridad.anio'] = pd.DatetimeIndex(csvfile['seguridad.fecha_hecho']).year
csvfile['seguridad.mes'] = pd.DatetimeIndex(csvfile['seguridad.fecha_hecho']).month
csvfile['seguridad.dia'] = pd.DatetimeIndex(csvfile['seguridad.fecha_hecho']).day
csvfile['seguridad.hora'] = pd.DatetimeIndex(csvfile['seguridad.fecha_hecho']).hour
csvfile['seguridad.dia_semana'] = pd.DatetimeIndex(csvfile['seguridad.fecha_hecho']).dayofweek

csvfile.head()

#csvfile.to_csv(r'E:\Maestria\prueba.csv', index = False)
```

# Agrupación de hurtos por año

```python
csvfile['seguridad.anio'].unique()
```

```python
datosAgrupados = csvfile.groupby(['seguridad.anio']).count()
```

# Gráfica de hurtos por año

```python
plt.plot(datosAgrupados.index, datosAgrupados['seguridad.fecha_hecho'])
plt.rcParams['figure.figsize'] = 10,6
plt.xlabel('año')
plt.ylabel('hurtos')
plt.xscale('linear')
plt.xticks(datosAgrupados.index)
plt.show
```

# Graficas de hurtos por dias en el año 2020

```python
anio2020 = csvfile[csvfile['seguridad.anio'] == 2020]
dias = ["Lunes","Martes","Miercoles","Jueves","Viernes","Sabado","Domingo"]
hurtosDias2020 = anio2020.groupby(['seguridad.dia_semana']).count()
```

```python
plt.barh(dias,hurtosDias2020['seguridad.fecha_hecho'])
plt.xlabel("Cantidad de hurtos")
plt.ylabel("Dias de la semana")
plt.title("Cantidad de hurtos en el año 2020 por dias de la semana")
plt.show()
```
